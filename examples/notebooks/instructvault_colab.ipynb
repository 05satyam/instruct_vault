```python
# If you're not on Colab, change this path to your local repo folder.
repo_path = "/content/ivault_demo"
print("Using repo_path:", repo_path)
```

```python
from pathlib import Path

repo = Path("/content/content/ivault_demo")
(repo / "prompts").mkdir(parents=True, exist_ok=True)

prompt_text = '''spec_version: "1.0"
name: hello_world
variables:
  required: [name]
messages:
  - role: system
    content: "You are a helpful assistant."
  - role: user
    content: "Say hello to {{ name }}."
tests:
  - name: includes_name
    vars: { name: "Ava" }
    assert: { contains_any: ["Ava"] }
'''

(repo / "prompts" / "hello_world.prompt.yml").write_text(prompt_text, encoding="utf-8")
print("Prompt written:", repo / "prompts" / "hello_world.prompt.yml")
```

```python
!ivault validate /content/content/ivault_demo/prompts
```

```python
from instructvault import InstructVault

vault = InstructVault(repo_root="/content/ivault_demo")
msgs = vault.render("prompts/hello_world.prompt.yml", vars={"name":"Colab"})
print(msgs)
```

```python
!ivault eval /ivault_demo/prompts/hello_world.prompt.yml --report /ivault_demo/out/report.json
```

```python
!ivault bundle --repo /content/ivault_demo --prompts prompts --out /content/ivault_demo/out/ivault.bundle.json
```
